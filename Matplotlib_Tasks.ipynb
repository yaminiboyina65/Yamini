```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
days = np.arange(1, 11)  
stock_price = [150, 152, 149, 155, 160, 162, 158, 165, 170, 175]  
trading_volume = [1.2, 1.4, 1.1, 1.8, 2.0, 2.5, 2.2, 2.7, 3.0, 3.5]
fig,axis1=plt.subplots()
axis1.plot(days,stock_price,color="red")
axis1.set_xlabel("Days")
axis1.set_ylabel("Stock Price")
axis2=axis1.twinx()
axis2.plot(days,trading_volume,color="blue")
axis2.set_ylabel("Trading Volume")
plt.title("Stock's Performance")
plt.tight_layout()
plt.show()
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
rent = [1200, 1200, 1200, 1200, 1250, 1250]
food = [400, 420, 450, 480, 500, 520]
entertainment = [150, 200, 220, 180, 250, 300]
transport = [100, 120, 130, 110, 140, 160]
plt.bar(months,rent,color="blue")
plt.bar(months,food,color="red")
plt.bar(months,entertainment,color="green")
plt.bar(months,transport,color="yellow")
plt.xlabel("Months")
plt.ylabel("Monthly expenses")
plt.title("Monthly expenses Comparison")
plt.legend(["Rent","Food","Entertainment","Transport"])
plt.show()
```

```python
weeks = np.arange(1, 9)  
covid_cases = [500, 700, 1200, 2000, 3000, 4500, 6000, 8000]  
vaccination_rate = [5, 10, 20, 35, 50, 65, 75, 85] 
fig,(axis1,axis2)=plt.subplots(2,1)
axis1.plot(weeks,covid_cases,color="red",linestyle="--",marker="o")
axis1.set_xlabel("Weeks")
axis1.set_ylabel("Covid_Cases")
axis1.set_title("Line Plot")
axis2.bar(weeks,vaccination_rate,color="green")
axis2.set_xlabel("Weeks")
axis2.set_ylabel("Vaccination_Rate")
axis2.set_title("Bar Graph")
plt.tight_layout()
plt.show()
```
